```python
import pandas as pd
import numpy as np
```

Date data

```python
# Dict that take 'date' as a key and define values as a list that contain date. 
df = pd.DataFrame({'date': ['2013-06-11','2014-05-08','2013-10-26','2015-08-20','2014-07-22']})
display(df)
print(df.dtypes)
```

```python
df['date'] = pd.to_datetime(df['date'])
print(df.dtypes) 
```

```python
# things you can do with this
df['dt_year'] = df['date'].dt.year
df['dt_month'] = df['date'].dt.month
df['dt_day'] = df['date'].dt.day
df['dt_quarter'] = df['date'].dt.quarter
df['dt_week'] = df['date'].dt.isocalendar().week 
df['dt_dayofweek'] = df['date'].dt.day_name()
```

Date + time data

```python
date = pd.Series(pd.date_range('2015-1-5 11:20:00', periods=7, freq='H'))

dt = pd.DataFrame(dict(date=date))
print(dt) 
```

```python
dt['time'] = dt['date'].dt.time
dt['hour'] = dt['date'].dt.hour
dt['min'] = dt['date'].dt.minute
dt['sec'] = dt['date'].dt.second
```

```python
date1 = pd.Series(pd.date_range('2012-1-1 00:00:00', periods=7, freq='S'))
date2 = pd.Series(pd.date_range('2012-1-1 00:00:00', periods=7, freq='H'))
 
time = pd.DataFrame(dict(Start_date = date1, End_date = date2))

time
```

```python
time['diff'] = time['End_date'] - time['Start_date']
time['diff_seconds']=time['diff']/np.timedelta64(1,'s')
time['diff_mins']=time['diff_seconds']/60
time['diff_hours']=time['diff_seconds']/3600

display(time)
```
