```python
import matplotlib.pyplot as plt

# sample from standard normal distribution 3000 times and save into an array of 3000 elements
samples_from_norm_dist = np.random.randn(3000)

# if we sample enough from the distribution, we see the distribution
# this is how standard normal distribution looks like
sns.histplot(samples_from_norm_dist)
plt.show()
```

```python
# the mean of the samples should be close to its original distribution at zero
np.mean(samples_from_norm_dist)
```

```python
# the std should be close to its orginical distrubution at one
np.std(samples_from_norm_dist)
```

```python
np.array([1,2,3,4]) + 2.5
```

```python
# we can customize our samples by multiplying the std and adding to the mean

# here let's increase std by 5
custom_arr = samples_from_norm_dist * 10

# observe the x axis, and see that the range is a lot wider compared to the sample distribution above
sns.histplot(custom_arr)
plt.show()
```

```python
# let's shift the mean by 20 by adding 20 to the array
custom_arr_2 = samples_from_norm_dist * 5 + 20

# observe the x axis, and see the mean shift
sns.histplot(custom_arr_2)
plt.show()
```
